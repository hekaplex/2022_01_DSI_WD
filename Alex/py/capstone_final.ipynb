Correlation Heatmap. High Correlation in number of affected people by natural disasters and number of homeless people; insured damages and total damages; aid contribution and number of homeless; number of injured and number of homeless

```python
plt.figure(figsize=(15,8))
sns.heatmap(nd.corr(), annot=True, cmap='Blues') # draws heatmap with input as correlation matrix calculated by iris.corr() 
plt.show()
```

How frequently do disasters occur worldwide

```python
disaster_types = nd["Disaster Type"].unique()
print("Types of Disasters reported:\n\n", disaster_types)
print("Occurrences:\n\n",nd["Disaster Type"].value_counts())
```

Years with most Natural Disasters

```python
d=nd['Year'].value_counts().reset_index()
d.head(10)
```

The rate of natural disasters in the world over years has increased almost proportionally with time. Plotted Below:

```python
sns.lmplot(x='index',y='Year',data=d.sort_values(by='Year'),
           aspect=1.7, height=5,markers=['o'], scatter_kws={'s':20});
```

Box plot. Rate of disasters by year and continent 

```python
sns.catplot(data=nd,
            kind='box', x='Year', y='Continent', orient='h', palette="Blues")

fig=plt.gcf()
fig.set_size_inches(20, 6)
plt.show()
```

The rate of natural disasters by type and rate of natural disasters by region. Flood, Storm, Erthquake, Epidemic have the highest rate. Souther and South-Eastern Asia, Eastern Asia, South America have the highest rate of Natural Disasters.
Plotted Below:

```python
fig, (axis1,axis2) = plt.subplots(2,1,figsize=(25,10))

axis1.set_title("Disaster Type")
axis2.set_title('Region')
plt.xticks(rotation=90,fontsize = 10)

sns.barplot(x='index',y='Disaster Type',ax=axis1, palette="Blues_r", data=nd['Disaster Type'].value_counts().reset_index())
sns.barplot(x='index',y='Region',ax=axis2, palette="Blues_r", data=nd['Region'].value_counts().reset_index())
```

Which Countries are prone to what type of Natural Disasters

```python
d=nd[['Country',"Disaster Type"]].groupby('Country').describe()
d["Disaster Type"].sort_values(by='count',ascending=False).head(10)
```

Which Countries are prone to what subtype of Natural Disasters

```python
d=nd[['Country',"Disaster Subtype"]].groupby('Country').describe()
d["Disaster Subtype"].sort_values(by='count',ascending=False).head(10)
```

Top 10 countries by number of Disasters

```python
p = nd[['Country',"Disaster Type"]].groupby('Country').count()
p.reset_index(inplace=True)
p=p.sort_values(by="Disaster Type",ascending=False).head(10)
p
```

Plotting All cases reported of natural disasters per country (Top 10)

```python
plt.xticks(rotation=90,fontsize = 10)
sns.barplot(x="Country", y="Disaster Type", data=p, palette="Blues_r")

fig=plt.gcf()
fig.set_size_inches(20, 4)
plt.show()
```

United States seems to be the most disaster prone country, Lets find out more:

US reports over time (Top 10 years by rate of Natural Disasters):

```python
n=nd.query('Country=="United States of America (the)"')['Year'].value_counts().reset_index()
n.head(10)
```

1991 and 1997 the worst years. Also most severe years are in the past 20 years range.
The rate of natural disasters in US  has increased almost proportionally with time. Plotted Below:

```python
fig, (axis1) = plt.subplots(figsize=(25,10))
plt.xticks(rotation=90)
sns.barplot(x='index',y='Year',data=n.sort_values(by='Year'), palette="Blues")
```

Total damages after Natural Disasters in United States. Plotted below:

```python
n=nd.query('Country=="United States of America (the)"')
```

```python
sns.relplot(data=n, kind='scatter',
            x='Year', y="Total Damages ('000 US$)")
fig=plt.gcf()
fig.set_size_inches(25, 6)
plt.show()
```

Let's find out which events caused the biggest damages

```python
n=nd.query('Country=="United States of America (the)"')[['Start Year','Start Month','Start Day',"Total Damages ('000 US$)"]]
n.reset_index().sort_values(by="Total Damages ('000 US$)",ascending=False)
```
