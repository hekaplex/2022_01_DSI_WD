```python
# clean Age
t[['Age']] = t[['Age']].fillna(30)
```

```python
# Create DPY
t['DollarPerYear'] = (t["Fare"] /t['Age'])
```

```python
#aggregate to avg DPY
t_agg = pd.DataFrame(
    {
        'Average DPY':
            t.groupby('Sex')
                .mean()['DollarPerYear']
    }
)
t_agg
```

```python
t_agg.to_csv("t_agg.csv")
```
